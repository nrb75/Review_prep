1. SQL command - selet AS

rename column that you are selecting. Select the Duration column, but rename it latency and keep it in a dataframe

```python
df.loc[:, ['Duration']].rename(columns={'Duration': 'latency'}).head()
```

2. SQL JOINs

left join - bring the right table into the left table, if something does not exist in left table the value from the right will be dropped. If the right table is missing data, it will simply be NaN

```python
df_left_join=df.merge(df_rules, on='pairs', how='left')
df_left_join.head()
```

Inner Join - merges the 2 tables and only retains values that both tables have in common

```python
df_inner_join = df.merge(df_rules, on='pairs', how='inner')
df_inner_join.head()
```

3. SQL GROUP BY functions

Group the rules_df by servers and calculate metrics like avg, count, etc.

```python
rules_mean=df_rules.groupby('server', as_index=False).mean()
rules_mean.head()
```

```python
#if you only want 1 column
rules_count=df_rules.groupby('server', as_index=False)['total_duration'].count()
rules_count

#df.groupby(['A','C'], as_index=False)['B'].sum()
```
