```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import MaxAbsScaler
import seaborn as sns
def minmax(var):
    res = []
    mi = min(var)  
    ma = max(var)
    for i in var:
        res.append((i-mi)/(ma-mi))
        
    return res
```

```python
taskDF = pd.read_csv('Task Statements.txt', delimiter='\t')
a = taskDF.groupby('O*NET-SOC Code')['Task'].apply(lambda x: "%s" % ' '.join(x))
a = a.to_frame()
persDF = pd.read_csv('Work Styles.txt', delimiter='\t')
persDF = persDF[persDF['Element Name'] == 'Leadership']
persDF = persDF.set_index(keys = 'O*NET-SOC Code')
persDF = persDF['Data Value'].to_frame()
df = persDF.merge(a, left_index=True, right_index=True)
```

```python
xTrain, xVal, yTrain, yVal = train_test_split(df['Task'], df['Data Value'], test_size=.2)
```

# Bag of Words
## Extracting Features

```python
mycv = CountVectorizer(stop_words='english')
```

```python
mycv.fit(xTrain)
xTrainBoW = mycv.transform(xTrain)
xValBoW = mycv.transform(xVal)
```

## Analyses

```python
myLR = LinearRegression()
myLR.fit(xTrainBoW, yTrain)
myLR.score(xValBoW, yVal)
```

```python
preds = myLR.predict(xValBoW)
sns.scatterplot(yVal, preds)
```

```python
xscaler = MaxAbsScaler()
xTrainBoWSca = xscaler.fit_transform(xTrainBoW)
xValBoWSca = xscaler.fit_transform(xValBoW)
yTrainSca = minmax(yTrain)
yValSca = minmax(yVal)
```

```python
myMLP = MLPRegressor((1000,1000,500,500), alpha = .002)
myMLP.fit(xTrainBoWSca, yTrainSca)
myMLP.score(xValBoWSca, yValSca)
```

```python
preds=myMLP.predict(xValBoWSca)
sns.scatterplot(yValSca, preds)
```

# TF-IDF
## Feature Extraction

```python
myTFIDF = TfidfVectorizer(stop_words='english')
myTFIDF.fit(xTrain)
xTrainTFIDF = myTFIDF.transform(xTrain)
xValTFIDF = myTFIDF.transform(xVal)
```

## Analyses

```python
myLR = LinearRegression()
myLR.fit(xTrainTFIDF, yTrain)
myLR.score(xValTFIDF, yVal)
```

```python
preds=myLR.predict(xValTFIDF)
sns.scatterplot(yVal, preds)
```

```python
xscaler = MaxAbsScaler()
xTrainTFIDFSca = xscaler.fit_transform(xTrainTFIDF)
xValTFIDFSca = xscaler.fit_transform(xValTFIDF)
yTrainSca = minmax(yTrain)
yValSca = minmax(yVal)
```

```python
myMLP = MLPRegressor((1000,1000,500,500), alpha = .002)
myMLP.fit(xTrainTFIDFSca, yTrainSca)
myMLP.score(xValTFIDFSca, yValSca)
```

```python
preds=myMLP.predict(xValTFIDFSca)
sns.scatterplot(yValSca, preds)
```
